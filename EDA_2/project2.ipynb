# Data Cleaning

```python
df=pd.read_csv(r"D:\\Data Science\\Data set\\healthcare_dataset_updated.csv")
df.columns = [col.replace(" ", "_") for col in df.columns]
df.dropna(inplace=True)
df['admission_date']=pd.to_datetime(df['admission_date'])
df['discharge_date']=pd.to_datetime(df['discharge_date'])
numeric_column = 'numeric_column'
if numeric_column in df.columns:
    df[numeric_column].fillna(df[numeric_column].mean(), inplace=True)
categorical_column = 'categorical_column'
if categorical_column in df.columns:
    df[categorical_column].fillna('Unknown', inplace=True)
df.drop_duplicates(inplace=True)
print(df.shape)
df.isna().sum()
df.info()
```

# Checking for any duplicates

```python
print(df.duplicated().sum())
```

# Statistical Analysis

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
stat_an=numeric_data.describe()
print(stat_an)
print(df.head(20))
print(df.tail())
plt.figure(figsize=(15, 10))
sns.boxplot(data=numeric_data)
plt.title('Boxplot of Numeric Data')
plt.xticks(rotation=90)
plt.show()
```

# Performing correlation

```python
correlation=numeric_data.corr()
print(correlation)
plt.figure(figsize=(12,8))
sns.heatmap(correlation,cmap='spring',annot=True,fmt='.2f',linewidths=0.5)
plt.title("correlation")
plt.show()
```

# Performing covariance

```python
covarience=numeric_data.cov()
print(covarience)
plt.figure(figsize=(12,8))
sns.heatmap(covarience,cmap='rainbow',annot=True,fmt='.2f',linewidths=0.5)
plt.title("covarience")
plt.show()
```

### 1.How does BMI compare for patients with different smoking statuses? 

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='smoking_status', y='bmi', data=df)
plt.title('Comparison of BMI Across Different Smoking Statuses')
plt.xlabel('Smoking Status')
plt.ylabel('BMI')
plt.show()
```

### 2.What is the difference in heart rate (heart_rate) between patients with high and low cholesterol levels?

```python
threshold = 200  
df['cholesterol_category'] = df['cholesterol_level'].apply(lambda x: 'High' if x > threshold else 'Low')
plt.figure(figsize=(10, 6))
sns.boxplot(x='cholesterol_category', y='heart_rate', data=df)
plt.title('Comparison of Heart Rate Between High and Low Cholesterol Levels')
plt.xlabel('Cholesterol Level')
plt.ylabel('Heart Rate')
plt.show()
```

### 3.How do patients with and without a family history of cardiovascular disease compare in terms of blood pressure and BMI?

```python
plt.figure(figsize=(12, 10))
plt.subplot(2, 3, 1)
sns.boxplot(x='family_history', y='blood_pressure_sys', data=df)
plt.title('Systolic Blood Pressure by Family History')
plt.subplot(2, 3, 2)
sns.boxplot(x='family_history', y='blood_pressure_dia', data=df)
plt.title('Diastolic Blood Pressure by Family History')
plt.subplot(2, 3, 3)
sns.boxplot(x='family_history', y='bmi', data=df)
plt.title('BMI by Family History')
plt.tight_layout()
plt.show()
```

### 4.How does smoking status impact average BMI and heart rate?

```python
avg_bmi_heart_rate = df.groupby('smoking_status')[['bmi', 'heart_rate']].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='smoking_status', y='bmi', data=avg_bmi_heart_rate, palette='coolwarm')
plt.title('Average BMI by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average BMI')
plt.subplot(1, 2, 2)
sns.barplot(x='smoking_status', y='heart_rate', data=avg_bmi_heart_rate, palette='coolwarm')
plt.title('Average Heart Rate by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average Heart Rate')
plt.tight_layout()
plt.show()
```

### 5.What is the average BMI for each exercise frequency category?



```python
df['exercise_frequency'] = df['exercise_frequency'].astype('category')
df['bmi'] = pd.to_numeric(df['bmi'], errors='coerce')
df.dropna(subset=['exercise_frequency', 'bmi'], inplace=True)
average_bmi_by_exercise = df.groupby('exercise_frequency')['bmi'].mean().reset_index()
print(average_bmi_by_exercise)
```

### 6.What is the distribution of exercise frequency?

```python
df['exercise_frequency'] = df['exercise_frequency'].astype('category')
plt.figure(figsize=(10, 6))
sns.countplot(x='exercise_frequency', data=df, palette='Set2')
plt.title('Distribution of Exercise Frequency')
plt.xlabel('Exercise Frequency')
plt.ylabel('Count')
plt.show()
```

### 7.How is the disease history distributed among patients?

```python
df['disease_history'] = df['disease_history'].astype('category')
plt.figure(figsize=(10, 6))
sns.countplot(x='disease_history', data=df, palette='Set2')
plt.title('Distribution of Disease History Among Patients')
plt.xlabel('Disease History')
plt.ylabel('Count')
plt.show()
```

### 8.Is there a significant difference in cholesterol levels between different age groups?


```python
df['cholesterol_level'] = pd.to_numeric(df['cholesterol_level'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
bins = [20, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['20-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
df.dropna(subset=['cholesterol_level', 'age_group'], inplace=True)
plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='cholesterol_level', data=df, palette='Set2')
plt.title('Cholesterol Level by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Cholesterol Level')
plt.show()
```

### 9.What is the gender distribution?

```python
gender_distribution = df['gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_distribution.index, y=gender_distribution.values, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
print(gender_distribution)
```

### 10.What is the distribution of patient ages?


```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df.dropna(subset=['age'], inplace=True)
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, bins=30, color='skyblue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
print(df['age'].describe())
```
