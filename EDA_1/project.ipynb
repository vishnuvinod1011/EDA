# Data Cleaning

```python
df=pd.read_csv(r"D:\\Data Science\\Data set\\updated_sales_customer_interaction_data.csv")
df.columns = [col.replace(" ", "_") for col in df.columns]
df.dropna(inplace=True)
df['DateTime']=pd.to_datetime(df['DateTime'])
numeric_column = 'numeric_column'
if numeric_column in df.columns:
    df[numeric_column].fillna(df[numeric_column].mean(), inplace=True)
categorical_column = 'categorical_column'
if categorical_column in df.columns:
    df[categorical_column].fillna('Unknown', inplace=True)
df.drop_duplicates(inplace=True)
print(df.shape)
df.isna().sum()
df.info()
```

# 1. What is the total revenue generated across all sales?

```python




total_revenue = df['TotalAmount'].sum()
print(f"Total Revenue: {total_revenue}")
```

# 2. Which product or service category has the highest sales volume?

```python



category_sales = df.groupby('Category')['TotalAmount'].sum()
top_category = category_sales.idxmax()
top_category_sales = category_sales.max()

print(f"The category with the highest sales volume is '{top_category}' with total sales of {top_category_sales}.")



```

# 4. What percentage of total sales comes from repeat customers?


```python



repeat_customers = df[df.duplicated('CustomerID', keep=False)]
repeat_sales = repeat_customers['TotalAmount'].sum()
percentage_repeat_sales = (repeat_sales / total_revenue) * 100
print(f"Percentage of total sales from repeat customers: {percentage_repeat_sales:.2f}%")

```

# 15. How does sales performance vary by day of the week or time of day?

```python


import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = ('D:\\Data Science\\Data set\\updated_sales_customer_interaction_data.csv')
data = pd.read_csv(file_path)

# Convert DateTime column to datetime format
data['DateTime'] = pd.to_datetime(data['DateTime'])

# Extract day of the week and hour of the day
data['DayOfWeek'] = data['DateTime'].dt.day_name()
data['HourOfDay'] = data['DateTime'].dt.hour

# Sales performance by day of the week
sales_by_day = data.groupby('DayOfWeek')['TotalAmount'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)

# Sales performance by hour of the day
sales_by_hour = data.groupby('HourOfDay')['TotalAmount'].sum()

# Plot sales by day of the week
plt.figure(figsize=(12, 6))
sales_by_day.plot(kind='bar', color='skyblue')
plt.title('Sales Performance by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot sales by hour of the day
plt.figure(figsize=(12, 6))
sales_by_hour.plot(kind='line', marker='o', color='orange')
plt.title('Sales Performance by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales')
plt.grid()
plt.tight_layout()
plt.show()

# Display summarized data
print("Sales by Day of the Week:")
print(sales_by_day)
print("\nSales by Hour of the Day:")
print(sales_by_hour)
```

# 17. How does the volume of customer interactions vary across months and weekdays? 


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = ('D:\\Data Science\\Data set\\updated_sales_customer_interaction_data.csv')
df = pd.read_csv(file_path)

# Ensure the DateTime column is in the correct format
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Extract the month and weekday
df['month'] = df['DateTime'].dt.month_name()
df['weekday'] = df['DateTime'].dt.day_name()

# Group by month and weekday, then count the number of interactions
heatmap_data = df.groupby(['month', 'weekday']).size().unstack()

# Reorder the weekdays for better visualization
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(columns=weekday_order)

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".0f", cbar_kws={'label': 'Number of Interactions'})
plt.title('Customer Interactions by Month and Weekday')
plt.ylabel('Month')
plt.xlabel('Weekday')
plt.tight_layout()
plt.show()

```

# 18. What is the correlation between sales figures, interaction frequency, and product categories?


```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = ('D:\\Data Science\\Data set\\updated_sales_customer_interaction_data.csv')
df = pd.read_csv(file_path)

# Ensure the DateTime column is in the correct format
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Aggregate data to calculate sales, interaction frequency, and group by product categories
category_summary = df.groupby('Category').agg(
    total_sales=('TotalAmount', 'sum'),
    interaction_count=('TransactionID', 'count'),
    avg_sales=('TotalAmount', 'mean')  # Example metric for additional insight
).reset_index()

# Calculate the correlation matrix
correlation_matrix = category_summary[['total_sales', 'interaction_count', 'avg_sales']].corr()

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Between Sales Figures, Interaction Frequency, and Product Categories')
plt.show()
```
